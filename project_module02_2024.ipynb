```python
data = pd.read_csv("heart.csv")
data.head()
```

Data preparation

```python
## we will be comparing rest of the parameters/columns present in the data with respect to precence or absece of heart disease
data['target'] = data.target.replace({1: "Disease", 0: "No_disease"})
data['sex'] = data.sex.replace({1: "Male", 0: "Female"})
data['cp'] = data.cp.replace({1: "typical_angina",
                          2: "atypical_angina",
                          3:"non-anginal pain",
                          4: "asymtomatic"})
data['exang'] = data.exang.replace({1: "Yes", 0: "No"})
data['slope'] = data.cp.replace({1: "upsloping",
                          2: "flat",
                          3:"downsloping"})
data['thal'] = data.thal.replace({1: "fixed_defect", 2: "reversable_defect", 3:"normal"})
```

Based on the plots above, answer the following questions:    
1. What is the percentage of Samples with Disease?

```python
# Count number of samples for every target class
target_counts = data['target'].value_counts()

# Get (%) of samples w/disease
percentage_disease = (target_counts['Disease'] / len(data)) * 100
print(f"Samples (%) w/Disease: {percentage_disease:.2f}%")
```

2. what are the 3 continous features that shows a singnficanct statistical differnce in distribution with respect to the precence and absence of the disease?

```python
from scipy.stats import ttest_ind

cont_features = ['age', 'chol', 'thalach', 'oldpeak', 'trestbps']

# Define splits
disease_group = data[data['target'] == 'Disease']
no_disease_group = data[data['target'] == 'No_disease']

# Perform t-tests
results = []
for feature in cont_features:
    t_stat, p_value = ttest_ind(disease_group[feature], no_disease_group[feature], equal_var=False)
    results.append((feature, p_value))

# Convert into pd.DataFrame | sort by p-value
results_df = pd.DataFrame(results, columns=['Feature', 'p_value']).sort_values(by='p_value')

# Print top 3 features w/smallest 'p'
print("Top 3 features:")
print(results_df.head(3))
```

3. Can we see a clear seperation in terms of the presence/absence of disease in the features obtained from pca and tsne plots?

- PCA Plot:

    - Seperating the 2 classes based on the features is difficult.
- t-SNE Plot:
    - Similar class points are grouped together. However, it is still hard to distinguish b/w the 2 classes.

4. What is the optimal number of principal components in our case?

```python
# Cumulative explained var
explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Find number of components for 95% var
optim_components = np.argmax(explained_variance >= 0.95) + 1
print(f"Optimal number of components: {optim_components}")
```

5. what are the continous features with the highest correation with each other?

```python
# Create corr-mat
corr_mat = numeric_data.corr()

# Masking tril (lower triangle)
mask = np.tril(np.ones(corr_mat.shape), k=0)
masked_corr = corr_mat.where(mask == 0)

# Finding features w/highest corr
highest_corr_pair = masked_corr.stack().idxmax()  # indices
highest_corr_value = masked_corr.stack().max()    # values
print(f"Feature pair w/highest corr: {highest_corr_pair} | Correlation: {highest_corr_value:.4f}")

# Viz the corr-mat
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, cmap='mako', fmt=".2f", cbar=True)
plt.title("Correlation Matrix of Cont Features", fontsize=16)
plt.show();
```

```python
# Extract columns with int and float types
numeric_columns = data.select_dtypes(include=["int64", "float64"]).columns

# Print the numeric columns
print(numeric_columns)
```

We will be analysing the dataset using the fact that weather the drink comes under the category tea or not

```python
data['Beverage_category'].unique()
```

```python
data['Tea'] = data['Beverage_category'].apply(lambda x: 1 if x == 'Tazo® Tea Drinks' else 0)
data = data.drop('Beverage_category', axis=1)
```

```python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(
    data,
    columns=['Beverage', 'Beverage_prep'],
    prefixes=['bev', 'bevp']
)
```

```python
data = data.replace({True: 1, False: 0})
```

```python
data = data.applymap(lambda x: np.float64(str(x).replace('%', '')))
```

```python
# Create a pie chart of the 'Tea' column also write your observation form the plot

tea_counts = data['Tea'].value_counts()
labels = ['Non-Tea', 'Tea']
colors = ['skyblue', 'mediumorchid']

# Plot pie-chart
plt.figure(figsize=(6, 6))
plt.pie(tea_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, textprops={'fontsize': 12})
plt.title('Distribution of Tea vs Non-Tea Beverages', fontsize=14)
plt.show()
```

```python
# perform pca on the data and plot the explained variace ratio, what is the optimal number of principal components in this case ?
from sklearn.preprocessing import StandardScaler

# Standardize
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# PCA
pca = PCA()
pca.fit(scaled_data)  # Fit to data

# Ploting 'explained variance ratio'
plt.figure(figsize=(7, 4))  # Create a small plot (to fit print's output too)
plt.plot(
    np.cumsum(pca.explained_variance_ratio_),
    marker='o', linestyle='--', color='mediumblue',
)
plt.title('Explained Variance Ratio by Principal Components', fontsize=16)
plt.xlabel('Number of Principal Components', fontsize=12)
plt.ylabel('Cumulative Explained Variance', fontsize=12)
plt.grid()
plt.show();

# Optim number of components
optim_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.95) + 1
print(f"Optimal number of principal components: {optim_components}")
```

```python
# visualise the principal components, choose the number of principal components based on the above plot. What is you observation from the plot?

pca_optim = PCA(n_components=39)
reduced_data = pca_optim.fit_transform(scaled_data)

# Visualize 1st 2 principal components
plt.figure(figsize=(10, 7))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.3)
plt.title('Visualization of First Two Principal Components', fontsize=16)
plt.xlabel('Component 1', fontsize=12)
plt.ylabel('Component 2', fontsize=12)
plt.grid()
plt.show();
```

Can't distinguish b/w classes as there exists no clustering. This may indicate that the provided data doesn't have "groups" (classes). (For the chosen dims)

```python
# plot the first 2 components of tsne, whats you observation from the plot?
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
tsne_results = tsne.fit_transform(reduced_data)

# Plot the 1st 2 t-SNE components
plt.figure(figsize=(10, 7))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1], alpha=0.3)
plt.title('t-SNE on 39 Principal Components', fontsize=16)
plt.xlabel('Component 1', fontsize=12)
plt.ylabel('Component 2', fontsize=12)
plt.grid()
plt.show();
```

There exists a grouping of sorts (clustering), this indicates that the data has some form of classes from the selected dims.
